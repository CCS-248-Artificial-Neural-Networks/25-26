# 1. Determine whether the student passes (1) or fails (0) based on:
a.	Hours studied
b.	Hours of sleep
c.	Hours of free time
Given weights w1 = 0.6, w2, 0.4, bias = -3
Step function with a threshold of 1 

Predict the output of the following inputs:
1.	(x1, x2) = (8, 7)
2.	(x1, x2) = (3, 4)


```python
import numpy as np


weights_p1 = np.array([0.6, 0.4])
bias_p1 = -3
threshold_p1 = 1

def step_function(weighted_sum, threshold):
    return 1 if weighted_sum >= threshold else 0

def predict_student_outcome(inputs, weights, bias, threshold):
    weighted_sum = np.dot(inputs, weights) + bias
    output = step_function(weighted_sum, threshold)
    return output, weighted_sum

inputs_p1_a = np.array([8, 7])
inputs_p1_b = np.array([3, 4])


weighted_sum_a, output_a = predict_student_outcome(inputs_p1_a, weights_p1, bias_p1, threshold_p1)
print(f"Inputs: (Hours Studied, Hours of Free Time) = (8, 7)")
print(f"Weighted Sum: ({inputs_p1_a[0]} * {weights_p1[0]}) + ({inputs_p1_a[1]} * {weights_p1[1]}) + {bias_p1} = {weighted_sum_a}")
print(f"Output: {output_a} (Pass)")

print("-" * 50)

# Calculate and print results for inputs (3, 4)
weighted_sum_b, output_b = predict_student_outcome(inputs_p1_b, weights_p1, bias_p1, threshold_p1)
print(f"Inputs: (Hours Studied, Hours of Free Time) = (3, 4)")
print(f"Weighted Sum: ({inputs_p1_b[0]} * {weights_p1[0]}) + ({inputs_p1_b[1]} * {weights_p1[1]}) + {bias_p1} = {weighted_sum_b}")
print(f"Output: {output_b} (Fail)")
```

# 2.	(20 points) Logic Gate Simulation. Given the following setup for a perceptron, compute its output and verify whether it acts as an AND gate.
Given weights w1 = 1, w2 = 1, bias = -1.5
Step function with a threshold of 0
Inputs	:
	(0,0) , (0,1), (1,0), (1,1)


```python
import numpy as np

weights_p2 = np.array([1, 1])
bias_p2 = -1.5
threshold_p2 = 0

def step_function_p2(weighted_sum, threshold):
    return 1 if weighted_sum > threshold else 0

def compute_logic_gate_output(inputs, weights, bias, threshold):
    weighted_sum = np.dot(inputs, weights) + bias
    output = step_function_p2(weighted_sum, threshold)
    return output


inputs_p2 = [(0, 0), (0, 1), (1, 0), (1, 1)]


print("Inputs (x1, x2) | AND Gate Output | Perceptron Output")
print("-" * 60)

for input_pair in inputs_p2:
    inputs_array = np.array(input_pair)
    output_perceptron = compute_logic_gate_output(inputs_array, weights_p2, bias_p2, threshold_p2)
    
    output_and = 1 if input_pair[0] == 1 and input_pair[1] == 1 else 0
    print(f"({input_pair[0]}, {input_pair[1]})       | {output_and}               | {output_perceptron}")

is_and_gate = all(compute_logic_gate_output(np.array(i), weights_p2, bias_p2, threshold_p2) == (1 if i[0] == 1 and i[1] == 1 else 0) for i in inputs_p2)
print(f"\nDoes the perceptron act as an AND gate? {'Yes' if is_and_gate else 'No'}")
```

# 3.	(60 points) Perceptron comparison (One vs All). Given the 3 perceptron, using the same input, compute the output and decided on the predicted class is the WINNER. If a tie is present, compare and get the highest weighted sum. 


Inputs = [0.5, -1, 2, 1, 0]
Perceptron A Configuration:
•	Weights: WA = [1.0, -0.5, 0.2, 0.1, 0.0]
•	BiasA: 0.2
Perceptron B
•	Weights: WB = [0.2, 0.2, 0.5, -0.4, 0.3]
•	BiasB: 0.0
Perceptron C
•	Weights: WC = [-0.3, -0.1, 0.4, 0.0, 0.2]
•	BiasC: -0.6


```python
import numpy as np


common_inputs_p3 = np.array([0.5, -1, 2, 1, 0])

weights_a = np.array([1.0, -0.5, 0.2, 0.1, 0.0])
bias_a = 0.2

weights_b = np.array([0.2, 0.2, 0.5, -0.4, 0.3])
bias_b = 0.0

weights_c = np.array([-0.3, -0.1, 0.4, 0.0, 0.2])
bias_c = -0.6

def activation_function_p3(weighted_sum):
    return 1 if weighted_sum > 0 else 0

def compute_perceptron_output(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias
    output = activation_function_p3(weighted_sum)
    return output, weighted_sum

output_a, weighted_sum_a = compute_perceptron_output(common_inputs_p3, weights_a, bias_a)
output_b, weighted_sum_b = compute_perceptron_output(common_inputs_p3, weights_b, bias_b)
output_c, weighted_sum_c = compute_perceptron_output(common_inputs_p3, weights_c, bias_c)


print(f"Common Inputs: {common_inputs_p3}")
print("-" * 50)

print(f"Perceptron A:")
print(f"  Weights: {weights_a}")
print(f"  Bias: {bias_a}")
print(f"  Weighted Sum: {weighted_sum_a:.2f}")
print(f"  Output: {output_a}")

print("-" * 50)

print(f"Perceptron B:")
print(f"  Weights: {weights_b}")
print(f"  Bias: {bias_b}")
print(f"  Weighted Sum: {weighted_sum_b:.2f}")
print(f"  Output: {output_b}")

print("-" * 50)

print(f"Perceptron C:")
print(f"  Weights: {weights_c}")
print(f"  Bias: {bias_c}")
print(f"  Weighted Sum: {weighted_sum_c:.2f}")
print(f"  Output: {output_c}")

print("-" * 50)


outputs = {'A': output_a, 'B': output_b, 'C': output_c}
winners = [key for key, value in outputs.items() if value == 1]

if len(winners) == 1:
    print(f"The winning perceptron is: Perceptron {winners[0]}")
elif len(winners) > 1:
    print(f"There is a tie between Perceptrons: {winners}")
    tied_sums = {
        'A': weighted_sum_a,
        'B': weighted_sum_b,
        'C': weighted_sum_c
    }
    highest_sum = -np.inf
    winner_name = None
    for name in winners:
        if tied_sums[name] > highest_sum:
            highest_sum = tied_sums[name]
            winner_name = name
    print(f"The tie is broken by the highest weighted sum.")
    print(f"Perceptron {winner_name} has the highest weighted sum of {highest_sum:.2f}.")
    print(f"The predicted class (winner) is Perceptron {winner_name}.")
else:
    print("No perceptron outputted a '1'. There is no winner.")
```
