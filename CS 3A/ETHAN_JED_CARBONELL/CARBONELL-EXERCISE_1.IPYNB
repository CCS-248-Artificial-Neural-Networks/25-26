```python
#activation function
def calculate_activation(inputs, weights, bias):
    weighted_sum = sum(i * w for i, w in zip(inputs, weights))
    return weighted_sum + bias
```

## Problem 1: Student Pass/Fail Prediction
Determine whether the student passes (1) or fails (0) based on:
- Hours studied
- Hours of sleep
- Hours of free time

Given:
- Weights: w1 = 0.6, w2 = 0.4
- Bias = -3
- Step function with threshold = 1

Test inputs:
1. (x1, x2) = (8, 7)
2. (x1, x2) = (3, 4)

```python
def perceptron_predict(inputs, weights, bias, threshold):
    activation_score = calculate_activation(inputs, weights, bias)

    print(f"Inputs: {inputs}")
    print(f"  Weighted Sum = ({inputs[0]} * {weights[0]}) + ({inputs[1]} * {weights[1]}) + ({bias}) = {activation_score:.2f}")

    if activation_score >= threshold:
        output = 1
        print(f"  Result: {activation_score:.2f} >= {threshold}, Output: 1 (Pass)\n")
    else:
        output = 0
        print(f"  Result: {activation_score:.2f} < {threshold}, Output: 0 (Fail)\n")
    return output

weights = (0.6, 0.4)
bias = -3
threshold = 1

input_1 = (8, 7)
input_2 = (3, 4)

prediction_1 = perceptron_predict(input_1, weights, bias, threshold)
prediction_2 = perceptron_predict(input_2, weights, bias, threshold)

print("Summary of Results")
print(f"For inputs {input_1}, the predicted output is: {prediction_1} (Pass)")
print(f"For inputs {input_2}, the predicted output is: {prediction_2} (Fail)")
```

## Problem 2: Logic Gate Simulation - AND Gate

Given:
- Weights: w1 = 1, w2 = 1
- Bias = -1.5
- Step function with threshold = 0

Test all possible binary inputs: (0,0), (0,1), (1,0), (1,1)
Verify if it acts as an AND gate

```python
def perceptron_predict(inputs, weights, bias, threshold):
    activation_score = calculate_activation(inputs, weights, bias)
    
    print(f"Inputs: {inputs}")
    print(f"  Weighted Sum = ({inputs[0]} * {weights[0]}) + ({inputs[1]} * {weights[1]}) + ({bias}) = {activation_score}")
    if activation_score >= threshold:
        output = 1
        print(f"  Result: {activation_score} >= {threshold}, Output: 1\n")
    else:
        output = 0
        print(f"  Result: {activation_score} < {threshold}, Output: 0\n")
    return output

weights = (1, 1)
bias = -1.5
threshold = 0
all_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
expected_and_outputs = [0, 0, 0, 1] 
perceptron_outputs = []

for i, current_input in enumerate(all_inputs):
    prediction = perceptron_predict(current_input, weights, bias, threshold)
    perceptron_outputs.append(prediction)

print("Verification")
print("Input (A, B) | Expected AND | Perceptron Output | Match?")
print("-" * 57)
for i, (inputs, expected, actual) in enumerate(zip(all_inputs, expected_and_outputs, perceptron_outputs)):
    match = "Yes" if expected == actual else "No"
    print(f"    {inputs}   |      {expected}       |        {actual}          | {match}")
```

## Problem 3: Perceptron Comparison (One vs All)

Given 3 perceptrons with the same input, compute outputs and determine the winner.
If tie occurs, compare weighted sums to get the highest.

Input: [0.5, -1, 2, 1, 0]

- Perceptron A: Weights = [1.0, -0.5, 0.2, 0.1, 0.0], Bias = 0.2
- Perceptron B: Weights = [0.2, 0.2, 0.5, -0.4, 0.3], Bias = 0.0  
- Perceptron C: Weights = [-0.3, -0.1, 0.4, 0.0, 0.2], Bias = -0.6

```python
def calculate_perceptron_score(name, inputs, weights, bias):
    final_score = calculate_activation(inputs, weights, bias)
    
    print(f"Perceptron {name}")
    calculation_str = " + ".join([f"({inputs[i]} * {weights[i]})" for i in range(len(inputs))])
    weighted_sum_only = final_score - bias # For print
    print(f"  Weighted Sum = {calculation_str}")
    print(f"  Score = {weighted_sum_only:.2f} + ({bias}) = {final_score:.2f}\n")

    return final_score

inputs = [0.5, -1, 2, 1, 0]

weights_A = [1.0, -0.5, 0.2, 0.1, 0.0]
bias_A = 0.2
weights_B = [0.2, 0.2, 0.5, -0.4, 0.3]
bias_B = 0.0
weights_C = [-0.3, -0.1, 0.4, 0.0, 0.2]
bias_C = -0.6

score_A = calculate_perceptron_score('A', inputs, weights_A, bias_A)
score_B = calculate_perceptron_score('B', inputs, weights_B, bias_B)
score_C = calculate_perceptron_score('C', inputs, weights_C, bias_C)

results = {'A': score_A, 'B': score_B, 'C': score_C}
winner_name = max(results, key=results.get)
winner_score = results[winner_name]

print("Final Comparison")
print(f"Perceptron A Final Score: {score_A:.2f}")
print(f"Perceptron B Final Score: {score_B:.2f}")
print(f"Perceptron C Final Score: {score_C:.2f}")
print(f"The WINNER is Perceptron '{winner_name}' with the highest score of {winner_score:.2f}.")
```
