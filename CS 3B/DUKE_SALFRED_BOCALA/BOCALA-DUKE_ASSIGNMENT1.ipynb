```python
print('---------- Number 1 ----------')
HoursStudied = [8, 7]
HoursSlept = [3, 4]
num1_bias = -3
num1_weights = [0.6, 0.4]
num1_threshold = 1
HoursStudied_output = 0
HoursSlept_output = 0

for i in range(len(num1_weights)):
    HoursStudied_output += HoursStudied[i] * num1_weights[i]
    HoursSlept_output += HoursSlept[i] * num1_weights[i]

HoursStudied_output += num1_bias
HoursSlept_output += num1_bias

print(f'Hours Studied with bias: {HoursStudied_output} ')
print(f"Hours Slept with bias: {round(HoursSlept_output, 1)}")

HoursSlept_activate = HoursSlept_output > num1_threshold
HoursStudied_activate = HoursStudied_output > num1_threshold
print('Hours Studied Activation', int(HoursStudied_activate))
print('Hours Slept Activation:', int(HoursSlept_activate))
```

```python
print('---------- Number 2 ----------')
input1 = [0, 0]
input2 = [0, 1]
input3 = [1, 0]
input4 = [1, 1]
num2_bias = -1.5
input1_output = 0 + num2_bias
input2_output = 0 + num2_bias
input3_output = 0 + num2_bias
input4_output = 0 + num2_bias
num2_threshold = 0
num2_weights = [1, 1]
for i in range(len(num2_weights)):
    input1_output += input1[i] * num2_weights[i]
    input2_output += input2[i] * num2_weights[i]
    input3_output += input3[i] * num2_weights[i]
    input4_output += input4[i] * num2_weights[i]

input1_activation = input1_output > num2_threshold
input2_activation = input2_output > num2_threshold
input3_activation = input3_output > num2_threshold
input4_activation = input4_output > num2_threshold

print(f'{input1} with bias: {input1_output}')
print(f'{input2} with bias: {input2_output}')
print(f'{input3} with bias: {input3_output}')
print(f'{input4} with bias: {input4_output}')
print(f'{input1}: {int(input1_activation)}')
print(f'{input2}: {int(input2_activation)}')
print(f'{input3}: {int(input3_activation)}')
print(f'{input4}: {int(input4_activation)}')
```

```python
print('---------- Number 3 ----------')
num3_inputs = [0.5, -1, 2, 1, 0]
weight_A = [1.0, -0.5, 0.2, 0.1, 0.0]
weight_B = [0.2, 0.2, 0.5, -0.4, 0.3]
weight_C = [-0.3, -0.1, 0.4, 0.0, 0.2]
num3_threshold = 0
bias_A = 0.2
bias_B = 0
bias_C = -0.6
output_A = 0 + bias_A
output_B = 0 + bias_B
output_C = 0 + bias_C

for i in range(len(num3_inputs)):
    output_A += num3_inputs[i] * weight_A[i]
    output_B += num3_inputs[i] * weight_B[i]
    output_C += num3_inputs[i] * weight_C[i]

activation_A = output_A > num3_threshold
activation_B = output_B > num3_threshold
activation_C = output_C > num3_threshold

print(f'Perceptron A with bias: {round(output_A, 1)}')
print(f'Perceptron B with bias: {round(output_B, 1)}')
print(f'Perceptron C with bias: {round(output_C, 2)}')
print(f'Perceptron A: {int(activation_A)}')
print(f'Perceptron B: {int(activation_B)}')
print(f'Perceptron C: {int(activation_C)}')
```
